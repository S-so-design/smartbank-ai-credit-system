```python
import pandas as pd

# Load CSV from Colab environment
df = pd.read_csv("F1_Student_Mock_Bank_Data.csv")

# Approval logic
def check_approval(row):
    income = row['Income ($)']
    spend = row['Total Spending ($)']
    if income >= 1500 and spend <= 0.8 * income:
        return "Approved"
    else:
        return "Declined"

# Apply logic
df['Decision'] = df.apply(check_approval, axis=1)

# Display results
df[['Customer ID', 'Name', 'Month', 'Income ($)', 'Total Spending ($)', 'Decision']]
```

```python
df.to_csv("F1_Credit_Score_Results.csv", index=False)

# To download the file in Colab:
from google.colab import files
files.download("F1_Credit_Score_Results.csv")
```
